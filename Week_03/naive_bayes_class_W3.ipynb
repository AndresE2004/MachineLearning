## Bloque 1. Importar librerías

En este bloque cargamos las herramientas necesarias para:
- cálculo numérico
- visualización
- construcción de un dataset tabular
- entrenamiento de Naive Bayes


```python
# BLOQUE 1: IMPORTAR LIBRERÍAS

# NumPy para arreglos numéricos.
import numpy as np

# Matplotlib para gráficas.
import matplotlib.pyplot as plt

# Pandas para crear el dataset de frutas.
import pandas as pd

# GaussianNB es la implementación de Gaussian Naive Bayes.
from sklearn.naive_bayes import GaussianNB

# load_iris carga el dataset Iris.
from sklearn.datasets import load_iris
```

### Interpretación del bloque 1

Aunque este bloque no produce un resultado analítico, sí define el marco de trabajo de toda la sesión.

**Qué debe quedar claro:**
- `NumPy` y `Pandas` organizan y transforman los datos.
- `Matplotlib` hará visible lo que el modelo está haciendo.
- `GaussianNB` es el clasificador principal de esta clase.
- `load_iris` aporta un dataset limpio y estándar, ideal para centrarse en la lógica del algoritmo.


## Bloque 2. Cargar el dataset Iris

Aquí cargamos Iris y separamos:
- `iris_X`: variables de entrada
- `iris_y`: clases

Se seleccionan dos variables para poder visualizar el problema en un plano.


```python
# BLOQUE 2: CARGAR IRIS

# Cargamos el dataset.
iris = load_iris()

# Seleccionamos solo dos variables para trabajar en 2D.
iris_X = iris.data[:, 0:2]

# Guardamos las clases.
iris_y = iris.target

# Revisamos dimensiones y nombres.
print("Shape de iris_X:", iris_X.shape)
print("Shape de iris_y:", iris_y.shape)
print("Clases:", iris.target_names)
print("Variables:", iris.feature_names)
```

### Interpretación del bloque 2

La salida típica muestra:
- `iris_X` con forma `(150, 2)`
- `iris_y` con forma `(150,)`

**Interpretación:**
- Hay 150 flores.
- Cada una se describe con 2 variables numéricas.
- Cada observación tiene una etiqueta de clase.

**Qué debe entender el estudiante:**
- Igual que en KNN, aquí también existe una estructura `X` y `y`.
- El modelo probabilístico necesita ejemplos etiquetados para estimar distribuciones y probabilidades por clase.
- Reducir a dos variables no significa perder la esencia del problema; significa hacerlo visible.


## Bloque 3. Visualizar Iris

Antes de entrenar el modelo, observamos la distribución del dataset.

Esto ayuda a:
- identificar separación entre clases
- detectar solapamiento
- anticipar la dificultad del problema


```python
# BLOQUE 3: VISUALIZAR IRIS

# Creamos figura.
_, ax = plt.subplots(figsize=(7, 5))

# Dibujamos la dispersión.
ax.scatter(
    iris_X[:, 0],
    iris_X[:, 1],
    c=iris_y,
    cmap='Set1',
    edgecolors='k'
)

# Agregamos contexto visual.
ax.set_title("Distribución del dataset Iris")
ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])

plt.show()
```

### Interpretación del bloque 3

Esta gráfica muestra cómo se organizan las flores según dos características.

**Qué observar:**
- Una clase puede quedar claramente separada.
- Otras dos pueden aparecer parcialmente mezcladas.
- Eso sugiere que algunas regiones serán más fáciles de clasificar que otras.

**Enlace con Naive Bayes:**
- Naive Bayes no mira vecinos como KNN; mira probabilidades por clase.
- Sin embargo, la distribución visual de los puntos sigue siendo importante, porque nos da intuición sobre cómo podrían comportarse esas probabilidades.


## Bloque 4. Crear y entrenar Gaussian Naive Bayes

Aquí entrenamos el clasificador probabilístico.

Recordatorio:
- Naive Bayes clasifica usando el **Teorema de Bayes**
- `GaussianNB` supone distribuciones normales para las variables numéricas
- También asume independencia condicional entre variables


```python
# BLOQUE 4: CREAR Y ENTRENAR GAUSSIANNB

# Creamos el modelo.
naive_bayes = GaussianNB()

# Entrenamos el modelo.
# Aquí estima probabilidades previas, medias y varianzas por clase.
naive_bayes.fit(iris_X, iris_y)

print("Modelo GaussianNB entrenado correctamente")
```

### Interpretación del bloque 4

En este bloque el modelo queda listo para calcular probabilidades de clase.

**Qué aprende el modelo:**
- qué tan frecuente es cada clase
- cuál es la media de cada variable en cada clase
- qué tanta dispersión tiene cada variable dentro de cada clase

**Diferencia conceptual con KNN:**
- KNN recuerda observaciones y compara distancias.
- Naive Bayes construye una representación probabilística del problema.

**Mensaje docente:**
Este es un excelente punto para mostrar que dos algoritmos pueden resolver clasificación de maneras completamente distintas.


## Bloque 5. Crear una malla del plano

Construimos una malla de puntos sobre todo el espacio para preguntarle al modelo qué clase asignaría en cada región.

Esto permite transformar un modelo probabilístico en una imagen interpretable.


```python
# BLOQUE 5: CREAR LA MALLA DEL PLANO

# Agregamos margen.
offset = 0.2

# Rangos del eje X.
x_min, x_max = iris_X[:, 0].min() - offset, iris_X[:, 0].max() + offset

# Rangos del eje Y.
y_min, y_max = iris_X[:, 1].min() - offset, iris_X[:, 1].max() + offset

# Malla de puntos.
x_grid, y_grid = np.meshgrid(
    np.linspace(x_min, x_max, 200),
    np.linspace(y_min, y_max, 200)
)

# Predecimos clase en cada punto del plano.
pred_grid = naive_bayes.predict(np.c_[x_grid.ravel(), y_grid.ravel()])

# Reorganizamos la salida.
pred_grid = pred_grid.reshape(x_grid.shape)

print("Shape de x_grid:", x_grid.shape)
print("Shape de y_grid:", y_grid.shape)
print("Shape de pred_grid:", pred_grid.shape)
```

### Interpretación del bloque 5

Este bloque amplía la mirada: ya no solo observamos las flores existentes, sino el **espacio completo de posibles observaciones**.

**Qué significan las dimensiones:**
- Una malla de `(200, 200)` representa 40.000 puntos del plano.
- El modelo predice una clase para cada uno de ellos.
- Esto permitirá ver visualmente cómo organiza el espacio.

**Valor didáctico:**
- El estudiante entiende que un clasificador no solo responde sobre datos ya vistos.
- También construye una lógica general para responder sobre datos nuevos.


## Bloque 6. Visualizar fronteras de decisión

Aquí convertimos la salida probabilística del modelo en una imagen:

- el fondo coloreado muestra la clase predicha
- los puntos reales muestran las flores observadas


```python
# BLOQUE 6: VISUALIZAR FRONTERAS DE DECISIÓN

# Creamos figura.
_, ax = plt.subplots(figsize=(7, 5))

# Pintamos el fondo según la clase predicha.
ax.pcolormesh(
    x_grid,
    y_grid,
    pred_grid,
    cmap="Set1",
    shading="auto"
)

# Superponemos los puntos reales.
ax.scatter(
    iris_X[:, 0],
    iris_X[:, 1],
    c=iris_y,
    edgecolors='k',
    cmap="Set1"
)

# Etiquetas visuales.
ax.set_title("Fronteras de decisión con Gaussian Naive Bayes")
ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])

plt.tight_layout()
plt.show()
```

### Interpretación del bloque 6

Esta gráfica muestra cómo Naive Bayes divide el espacio entre clases.

**Qué debe notarse:**
- Las regiones suelen ser más suaves o regulares que en KNN.
- Eso ocurre porque la decisión se basa en probabilidades derivadas de distribuciones, no en vecinos puntuales.
- El fondo indica la clase más probable en cada región.

**Mensaje comparativo con KNN:**
- KNN produce fronteras guiadas por la cercanía local.
- Naive Bayes produce fronteras guiadas por la estructura probabilística estimada.

Esta comparación es didácticamente muy valiosa porque muestra dos formas de “pensar” un mismo problema.


## Bloque 7. Crear un segundo dataset: frutas

Ahora construimos un ejemplo más cotidiano para mostrar que el algoritmo puede aplicarse a otro contexto.

Las variables serán:
- `weight`
- `color_score`

La clase será el tipo de fruta.


```python
# BLOQUE 7: CREAR DATASET DE FRUTAS

# Creamos una tabla con nombre, peso y color_score.
fruit_data = pd.DataFrame({
    "fruit_name": [
        "apple", "apple", "apple", "apple",
        "orange", "orange", "orange", "orange",
        "mandarin", "mandarin", "mandarin", "mandarin"
    ],
    "weight": [150, 170, 160, 180, 140, 155, 165, 175, 80, 85, 90, 95],
    "color_score": [0.72, 0.76, 0.74, 0.78, 0.88, 0.90, 0.86, 0.89, 0.60, 0.58, 0.62, 0.61]
})

# Diccionario para convertir nombre en número.
fruit_map = {"apple": 0, "orange": 1, "mandarin": 2}

# Creamos la etiqueta numérica.
fruit_data["fruit_label"] = fruit_data["fruit_name"].map(fruit_map)

# Mostramos la tabla.
fruit_data
```

### Interpretación del bloque 7

Este bloque cumple dos funciones:

1. Mostrar que el algoritmo no depende exclusivamente de Iris.
2. Traducir un problema abstracto a uno más cotidiano.

**Qué debe entenderse:**
- Los modelos de clasificación necesitan clases codificadas.
- Por eso se transforma el nombre de la fruta a una etiqueta numérica.
- El dataset sigue teniendo la misma estructura básica: variables de entrada y una clase objetivo.

**Valor pedagógico:**
Este segundo ejemplo suele hacer que el estudiante se apropie mejor del problema porque le resulta más cercano.


## Bloque 8. Preparar variables del dataset de frutas

Aquí repetimos el patrón:
- `fruits_X` como variables de entrada
- `fruits_y` como etiquetas

También convertimos estas estructuras a arreglos NumPy.


```python
# BLOQUE 8: PREPARAR FEATURES Y LABELS DE FRUTAS

# Variables de entrada.
fruits_X = fruit_data[["weight", "color_score"]]

# Variable objetivo.
fruits_y = fruit_data["fruit_label"]

# Convertimos a arreglos NumPy.
fruits_X = np.asarray(fruits_X)
fruits_y = np.asarray(fruits_y)

print("Shape de fruits_X:", fruits_X.shape)
print("Shape de fruits_y:", fruits_y.shape)
```

### Interpretación del bloque 8

La salida muestra la dimensión del nuevo problema.

**Interpretación:**
- Cada fruta tiene 2 variables numéricas.
- Cada fruta tiene 1 etiqueta de clase.
- El patrón de preparación del problema es el mismo que en Iris.

**Lección importante:**
La estructura de un problema de clasificación suele mantenerse estable aunque cambie el dominio:
- entradas
- salidas
- entrenamiento
- predicción


## Bloque 9. Entrenar Naive Bayes con frutas

Ahora usamos el mismo algoritmo en el nuevo dataset.

Este paso refuerza la idea de que un modelo puede reutilizarse en distintos problemas.


```python
# BLOQUE 9: ENTRENAR NAIVE BAYES CON FRUTAS

# Creamos un nuevo modelo.
naive_bayes_fruits = GaussianNB()

# Entrenamos con el dataset de frutas.
naive_bayes_fruits.fit(fruits_X, fruits_y)

print("Modelo de frutas entrenado correctamente")
```

### Interpretación del bloque 9

Este bloque muestra que el algoritmo es reutilizable.

**Qué significa:**
- La misma lógica probabilística ahora se aplica a peso y color de frutas.
- El modelo estima cómo se distribuyen estas variables dentro de cada tipo de fruta.

**Mensaje docente:**
Una gran parte del Machine Learning consiste justamente en aplicar estructuras similares a contextos distintos.


## Bloque 10. Visualizar el dataset de frutas

Representamos las frutas en el plano:
- eje X: peso
- eje Y: color

Además escribimos el nombre de cada fruta junto a su punto.


```python
# BLOQUE 10: VISUALIZAR EL DATASET DE FRUTAS

# Creamos figura.
_, ax = plt.subplots(figsize=(7, 5))

# Dibujamos los puntos.
ax.scatter(
    fruits_X[:, 0],
    fruits_X[:, 1],
    c=fruits_y,
    cmap="Set2",
    edgecolors="k",
    s=120
)

# Título y etiquetas.
ax.set_title("Distribución del dataset de frutas")
ax.set_xlabel("Weight")
ax.set_ylabel("Color score")

# Escribimos el nombre de cada fruta.
for i, name in enumerate(fruit_data["fruit_name"]):
    ax.annotate(name, (fruits_X[i, 0] + 1, fruits_X[i, 1] + 0.003), fontsize=8)

plt.show()
```

### Interpretación del bloque 10

Esta visualización hace mucho más intuitivo el problema.

**Qué debe observarse:**
- Algunas frutas ocupan regiones bastante definidas del plano.
- Peso y color parecen suficientes para separar razonablemente varias clases.
- Etiquetar cada punto con el nombre facilita la interpretación en una clase magistral.

**Conclusión didáctica:**
Visualizar no es un adorno; es una forma de hacer visible la estructura del problema antes de confiar en el modelo.


## Bloque 11. Fronteras de decisión en frutas

Repetimos la lógica de la malla para observar cómo Naive Bayes divide el espacio en este nuevo problema.


```python
# BLOQUE 11: FRONTERAS DE DECISIÓN EN FRUTAS

# Márgenes del plano.
offset_x = 5
offset_y = 0.05

# Rangos.
x_min_f, x_max_f = fruits_X[:, 0].min() - offset_x, fruits_X[:, 0].max() + offset_x
y_min_f, y_max_f = fruits_X[:, 1].min() - offset_y, fruits_X[:, 1].max() + offset_y

# Malla.
x_grid_f, y_grid_f = np.meshgrid(
    np.linspace(x_min_f, x_max_f, 200),
    np.linspace(y_min_f, y_max_f, 200)
)

# Predicción sobre la malla.
pred_grid_f = naive_bayes_fruits.predict(np.c_[x_grid_f.ravel(), y_grid_f.ravel()])

# Reorganizamos.
pred_grid_f = pred_grid_f.reshape(x_grid_f.shape)

# Graficamos.
_, ax = plt.subplots(figsize=(7, 5))

ax.pcolormesh(
    x_grid_f,
    y_grid_f,
    pred_grid_f,
    cmap="Set2",
    shading="auto"
)

ax.scatter(
    fruits_X[:, 0],
    fruits_X[:, 1],
    c=fruits_y,
    cmap="Set2",
    edgecolors="k",
    s=120
)

ax.set_title("Fronteras de decisión en el dataset de frutas")
ax.set_xlabel("Weight")
ax.set_ylabel("Color score")

plt.tight_layout()
plt.show()
```

### Interpretación del bloque 11

Esta gráfica permite ver cómo el modelo generaliza más allá de las frutas observadas.

**Cómo leerla:**
- El fondo coloreado indica qué fruta sería la más probable en cada zona del plano.
- Los puntos reales muestran dónde están las frutas conocidas.
- Las regiones delimitan el criterio probabilístico del modelo.

**Valor docente:**
Este bloque ayuda a consolidar la idea de que el modelo ya no depende solo de los ejemplos exactos; ahora tiene una estructura que le permite clasificar nuevas observaciones.


## Bloque 12. Hacer una predicción nueva

Por último, creamos una fruta ficticia y consultamos:
- qué clase predice el modelo
- con qué probabilidad lo hace


```python
# BLOQUE 12: HACER UNA PREDICCIÓN NUEVA

# Nueva fruta ficticia.
nueva_fruta = np.array([[168, 0.87]])

# Clase predicha.
prediccion = naive_bayes_fruits.predict(nueva_fruta)[0]

# Probabilidades por clase.
probabilidades = naive_bayes_fruits.predict_proba(nueva_fruta)[0]

# Diccionario inverso para recuperar nombres.
inverse_fruit_map = {v: k for k, v in fruit_map.items()}

print("Nueva fruta:", nueva_fruta)
print("Clase predicha:", prediccion)
print("Nombre predicho:", inverse_fruit_map[prediccion])
print("Probabilidades por clase:")
for idx, prob in enumerate(probabilidades):
    print(f"  {inverse_fruit_map[idx]}: {prob:.4f}")
```

### Interpretación del bloque 12

Este bloque conecta el modelo con una decisión concreta.

**Cómo interpretar la salida:**
- La clase predicha es la más probable dadas las características de la fruta.
- Las probabilidades muestran cuánta evidencia encontró el modelo a favor de cada clase.
- Si una probabilidad domina claramente, la decisión es más fuerte.
- Si varias probabilidades son similares, la clasificación es más incierta.

**Comparación con KNN:**
- En KNN la probabilidad se relaciona con vecinos cercanos.
- En Naive Bayes la probabilidad se relaciona con el modelo estadístico de cada clase.
